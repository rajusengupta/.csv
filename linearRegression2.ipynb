```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as sstats
import random
from sklearn.model_selection import train_test_split

x = random.random()
print(x)
y = np.linspace(0,50,150)
print(y)

def f_y(a):
  return random.random()*a

def t_y(a):
  return 2*a

z1 = list(map(f_y, y))
print(z1)

z2 = list(map(t_y, y))
print(z2)

plt.plot(z1, z2, y, label = 'randomness2')
plt.legend()
plt.show()

slope, intercept, r, p, std_error = sstats. linregress(y, z1)
print(slope,' ', intercept,' ', r,' ', p,' ', std_error)

def lin_x(x):
  return slope*x + intercept

zs = list (map(lin_x, y))
plt.scatter(y, z1)
plt.plot(y, zs)
plt.show()

slope2, intercept2, r2, p2, std_error2 = sstats. linregress(y, z2)
print(slope2,' ', intercept2,' ', r2,' ', p2,' ', std_error2)

def lin_x2(x):
  return slope2*x + intercept2

zt = list (map(lin_x2, y))
plt.scatter(y, z2)
plt.plot(y, zt)
plt.show()

slope3, intercept3, r3, p3, std_error3= sstats. linregress(z1, z2)
print(slope3,' ', intercept3,' ', r3,' ', p3,' ', std_error3)

def lin_x3(x):
  return slope3*x + intercept3

zu = list (map(lin_x3, z1))
plt.scatter(z1, z2)
plt.plot(z1, zu)
plt.show()

Xtrain, Xtest, ytrain, ytest = train_test_split(z2, z1, train_size=0.55, test_size=0.45, random_state=42)
print('Xtrain: ', Xtrain)
print('Xtest: ', Xtest)
print('ytrain: ', ytrain)
print('ytest: ', ytest)

xtSlope,xtIntercept, xtR, xtP, xtStd_Err = sstats.linregress(Xtrain, ytrain)
xtsSlope,xtsIntercept, xtsR, xtsP, xtsStd_Err = sstats.linregress(Xtest, ytest)

def new1(x):
  return xtSlope*x + xtIntercept

def new2(x):
  return xtsSlope*x + xtsIntercept

z2trn= list((map(new1, Xtrain)))
z2test = list(map(new2,Xtest))

plt.scatter(Xtrain,ytrain)
plt.plot(Xtrain,z2trn)
plt.scatter(Xtest,ytest)
plt.plot(Xtest,z2test)
plt.show()

z2three = list(map(new1, Xtest))

plt.scatter(Xtrain,ytrain)
plt.plot(Xtrain,z2trn)
plt.scatter(Xtest,ytest)
plt.plot(Xtest,z2three)
plt.show()
```
